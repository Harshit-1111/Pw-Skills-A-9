```python
try:
    result = 10 / 0 
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
'''Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.'''


The LookupError class in Python is a base class for exceptions that occur when attempting to access or
look up elements in a collection (such as a list, dictionary, or tuple) and the specified key or index is not found or is invalid.
It serves as the parent class for exceptions like KeyError and IndexError, both of which represent specific lookup-related errors. 
```

```python
my_dict = {"apple": 3, "banana": 5, "cherry": 2}

try:
    value = my_dict["grape"] 
except KeyError as e:
    print(f"Error: {e}")
```

```python
my_list = [1, 2, 3, 4, 5]

try:
    value = my_list[10]  
except IndexError as e:
    print(f"Error: {e}")
```

```python
'''Q5. Explain ImportError. What is ModuleNotFoundError?'''

ImportError:

ImportError is a base class for exceptions that occur when there is a problem with importing a module. 
It can be raised in various situations, including when a module cannot be found or when there is an issue with the module's content.


ModuleNotFoundError:

ModuleNotFoundError is a more specific exception introduced in Python 3.6. 
It is raised when an import statement fails to locate the specified module. It is a subclass of ImportError.
Unlike ImportError, ModuleNotFoundError explicitly indicates that the module being imported could not be found,
making it easier to diagnose import-related issues.
```

```python
try:
    import non_existent_module  
except ImportError as e:
    print(f"Import Error: {e}")
```

```python
try:
    import non_existent_module 
except ModuleNotFoundError as e:
    print(f"Module Not Found Error: {e}")
```
